#### 基本使用

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])   
# 如果传入单个列表或数组，则plt会默认这是y的值，并给一组默认的x坐标
# plot中有第三个参数，指定颜色和样式，默认为"b-"是一条蓝色实线
plt.ylabel('some numbers')
plt.show()
```

```python
plt.plot([1, 4, 9, 16], 'ro')    # o 是样式描述，表示圆点
plt.ylabel('some numbers')
plt.show()
```

```python
# 传入numpy的数组
import numpy as np
```

```python
x = np.arange(2,23)
y_1 = x**1.3
y_2 = x**2

plt.plot(x, x, 'r--', x, y_1, 'bs', x, y_2, 'g^')
plt.show()
```

#### 多个图形和轴域

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure(1)
plt.subplot(211)   # subplot()命令指定numrows，numcols，fignum，其中fignum的范围是从1到numrows * numcols
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(2, 1, 2)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

```python
plt.figure(1)                # 第一个图形
plt.subplot(211)             # 第一个图形的第一个子图
plt.plot([1, 2, 3])
plt.subplot(212)             # 第一个图形的第二个子图
plt.plot([4, 5, 6])


plt.figure(2)                # 第二个图形
plt.plot([4, 5, 6])          # 默认创建 subplot(111)

plt.figure(1)                # 当前是图形 1，subplot(212)
plt.subplot(211)             # 将第一个图形的 subplot(211) 设为当前子图
plt.title('Easy as 1, 2, 3') # 子图 211 的标题
```

####  添加文本

```python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# 数据的直方图
n, bins, patches = plt.hist(x, 50, normed=1, facecolor='g', alpha=0.75)


plt.xlabel('Smarts', fontsize=14, color='red')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```
