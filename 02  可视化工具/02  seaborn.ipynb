```python
import seaborn as sns

sns.__version__
```

#### 简介部分

Seaborn 是一个基于 matplotlib 且数据结构与 pandas 统一的统计图制作库。

```python
# 设置并使用 seaborn 默认的主题、尺寸大小以及调色板
%matplotlib inline
sns.set()
```

```python
# 官方示例，这里 load_dataset 会导入一个类似 pandas.DataFrame 的数据结构
# replot 可以绘制 散点图 和 折线图 ，用于展示数据之间的关系，默认为散点图（参数 kind='scatter'）

# This function provides access to several different axes-level functions that show the relationship
# between two variables with semantic mappings of subsets.

# 下图展示出 tips 数据集中五个变量的关系，三个是数值变量，另外两个是类别变量

tips = sns.load_dataset("tips")
sns.relplot(x="total_bill", y="tip", col="time",
            hue="smoker", style="smoker", size="size",
            data=tips)
```

```python
# 可以非常简单地调用
sns.relplot(x="total_bill", y="tip",
            hue="smoker", style="smoker", size="size",
            data=tips)
```

```python
# 如果有一个变量具有时间意义，线状图可能更好一点
dots = sns.load_dataset("dots")
sns.relplot(x="time", y="firing_rate", col="align",
            hue="choice", size="coherence", style="choice",
            facet_kws=dict(sharex=False),
            kind="line", legend="full", data=dots);
```

####  相关关系的可视化：relplot()、scatterplot()、lineplot()

```python
# 以之前的数据集为例，replot默认以散点图的方式呈现数据的类别

sns.relplot(x="total_bill", y="tip", hue="smoker", style="smoker",
            data=tips)
```

```python
# 类别同样可以使用自定义的调色板

sns.relplot(x="total_bill", y="tip", hue="size", palette="ch:r=-.5,l=.75", data=tips)
```

调色板的教程可以参考[这篇文章](https://www.jianshu.com/p/26c72ba991af)

```python
# size 用来表示每个点的大小，需要注意的是，这里圆点的面积并不是由size这列数据指定的
sns.relplot(x="total_bill", y="tip", size="size", data=tips)
```

```python
sns.relplot(x="total_bill", y="tip", size="size", sizes=(15, 200), data=tips)
```
