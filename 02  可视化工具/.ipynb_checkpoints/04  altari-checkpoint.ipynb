```python
import altair as alt
```

```python
# 这里先使用sklearn.datasets中的diabetes数据集，这个数据集中特征列均经过处理，

from sklearn import datasets
 
X,y = datasets.load_diabetes(return_X_y=True)
```

```python
import pandas as pd
```

```python
X_form = pd.DataFrame(X, columns=['age', 'sex', 'mass_index', 'avg_blood_press', 'S1', 'S2', 'S3', 'S4', 'S5', 'S6'])
y_form = pd.DataFrame(y, columns=['outcome'])
```

```python
alt.Chart((pd.concat([X_form, y_form], axis=1))).mark_point().encode(
    x='mass_index',
    y='S1',
    color = 'sex:N'
)
```
