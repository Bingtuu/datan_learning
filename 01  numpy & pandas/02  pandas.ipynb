#### 索引与数据选择

```python
# DataFrame可以通过传入列名称直接选取数据，这是基本索引
data[['adr','customer_type']].head()
```

```python
# 也可以通过属性直接访问列
data.lead_time
```

```python
# loc是基于名称的索引，先行后列地指定行和列的索引名称
data.loc[5:120:2]   # 传入行索引，列默认全选
```

```python
# loc的另一种用法
data.loc[5:10, ['agent', 'company']]
```

```python
data.loc[5:10, 'hotel': 'company']
```

```python
# iloc是基于位置的索引,先行后列地指定行和列的位置 
data.iloc[:,3:5].head()
```

```python
# 可以单取一行数据
data.iloc[2]
```

```python
data.iloc[2:100, :].head()
```

```python
# []、.loc、.iloc 可以接收一个“索引器”
data[data['adr'] > data['adr'].mean()]
```

```python
data.loc[lambda data : data.adr > 70, :]
```

```python
# 重建索引
data_reindex = data.copy()
data_reindex.reindex([i for i in range(1, len(data_reindex) + 1)])
```

```python
# 随机采样
data.sample(n=5, random_state=101)
```

```python
# 创建一列新数据
data.loc[:, 'tags'] = data['adults'] + data['children']
data.loc[:, 'tags']
```

```python
# .at 和 .iat 可以迅速找到数据
data.at[5,'adults']
```

```python
# 布尔索引
data[(data['adults'] > 2) | (data['children'] >4)]  # 条件之间的括号是必须的（0.25），否则这里断句会出现问题导致报错
```

```python
data[(data['adults'] > 1) & (data['children'] >3)]
```

```python
data[~(data['children'] > 2)]
```

```python
# .isin() 索引
data[data['distribution_channel'].isin(['Direct', 'Corporate'])]
```

```python
# where()
data['adults'].where(data['adults'] >= 2)
```

```python
# mask
data['adults'].mask(data['adults'] >= 2)
```

```python
# 利用 where() 和 mask() 对数据进行处理
data_where  = data.copy()
data_where['adults'] = data_where['adults'].where(data['adults'] >= 2, 2)
data_where['adults']
```

```python
data_mask  = data.copy()
data_mask['adults'] = data_mask['adults'].mask(data['adults'] >= 2, 2)
data_mask['adults']
```

```python
# query() 方法
data.query('adults >=2 and children == 2')
```

```python
data.query('adults <= children')
```

```python
data.query('["Direct", "Corporate"] not in distribution_channel')  # 等同于 ' distribution_channel == ["Direct", "Corporate"]'
```

####  数据合并

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']},
                   index=[0, 1, 2, 3])
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'], 
                    'B': ['B0', 'B1', 'B2', 'B3'], 
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']},
                   index=[0, 1, 2, 3])
df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                    'B': ['B8', 'B9', 'B10', 'B11'],
                    'C': ['C8', 'C9', 'C10', 'C11'],
                    'D': ['D4', 'D5', 'D6', 'D7']},
                   index=[8, 9, 10, 11])

frames = [df1, df2, df3]
result = pd.concat(frames)
result                   
```

```python
# 在 axis=1 横向拼接时，concat 会按照索引进行，这里ignore_index 的含义为是否保留原来的列名称作为索引
result2 = pd.concat(frames, axis=1, ignore_index=True)
result2
```

```python
# 可以通过 key 来指定拼接的 DataFrame 的名称
result3 = pd.concat(frames, keys=['x', 'y', 'z'])
result3
```

```python
result3.loc['y']
```

```python
# 另一个参数 join 用来指定连接的行为，这里注意 df4 的 index

df4 = pd.DataFrame({'B': ['B2', 'B3', 'B6', 'B7'],
                    'D': ['D2', 'D3', 'D6', 'D7'],
                    'F': ['F2', 'F3', 'F6', 'F7']},
                   index=[2, 3, 6, 7])
result4 = pd.concat([df1, df4], axis=1, sort=True)
result4
```

```python
result5 = pd.concat([df1, df4], axis=1, join='inner', sort=False)
result5
```

```python
result6 = pd.concat([df1, df4.reindex(df1.index)], axis=1)
result6
```

```python
# append 默认根据列进行拼接，这样可以将相同类名称（column name）进行数据对齐
result = df1.append(df2)
result
```

```python
# 注意没有指定 sort 参数时的提示信息（0.25 版本）
result2 = pd.concat([df1, df4], ignore_index=True)
result2
```

```python
result3 = df1.append(df4, sort=False)
result3
```

```python
# merge 相当于 SQL 中的连表，需要用参数 on 指定关联的 key ，如果有多个需要传入一个列表
# on 参数有 left、right、inner、outer
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})

right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                      'key2': ['K0', 'K0', 'K0', 'K0'],
                      'C': ['C0', 'C1', 'C2', 'C3'],
                      'D': ['D0', 'D1', 'D2', 'D3']})

result = pd.merge(left, right, on=['key1', 'key2'])
result
```

```python
# 注意 validate 参数的作用，，默认为 None，可以选择 one_to_one、one_to_many、many_to_one、many_to_many
left = pd.DataFrame({'A' : [1,2], 'B' : [1, 2]})
right = pd.DataFrame({'A' : [4,5,6], 'B': [2, 2, 2]})

pd.merge(left, right, on='B', how='outer', validate="one_to_many")
```

```python
# 可以使用 indicator=True 来显示 _merge 参数
df1 = pd.DataFrame({'col1': [0, 1], 'col_left': ['a', 'b']})
df2 = pd.DataFrame({'col1': [1, 2, 2], 'col_right': [2, 2, 2]})

pd.merge(df1, df2, on='col1', how='outer', indicator=True)
```

```python
# join 是 DataFrame 的方法
left = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                    index=['K0', 'K1', 'K2'])

right = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                      'D': ['D0', 'D2', 'D3']},
                     index=['K0', 'K2', 'K3'])

result = left.join(right)
result
```

```python
result2 = left.join(right, how='outer')
result2
```

```python
left = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3'],
                     'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1']})

index = pd.MultiIndex.from_tuples([('K0', 'K0'), ('K1', 'K0'),
                                   ('K2', 'K0'), ('K2', 'K1')])

right = pd.DataFrame({'C': ['C0', 'C1', 'C2', 'C3'],
                      'D': ['D0', 'D1', 'D2', 'D3']},
                     index=index)

result4 = left.join(right, on=['key1', 'key2'])

result4
```

####  样式

```python
data = pd.read_excel(r'datasets\style_sample.xlsx')
data
```

pandas.style 的 api ：https://pandas.pydata.org/pandas-docs/stable/reference/style.html

pandas的样式调整常用的有两种情况：
* `Styler.applymap` 用于调整元素的样式
* `Styler.apply` 用于调整表格、行列的样式

```python
# 定义一个让负数突出显示为红色的函数
def color_neg_red(x):
    color = 'red' if x < 0 else 'black'
    return 'color:{color_name}'.format(color_name=color)

a = data.style.applymap(color_neg_red)
a
```

```python
# 将每一类的最大值突出显示
def hilight_max(x):
    is_max = x == x.max()
    return ['background-color: yellow' if v else '' for v in is_max]

b = data.style.apply(hilight_max)
b
```

```python
# 重叠两种样式

data.style.applymap(color_neg_red).apply(hilight_max)
```

```python
# 对指定的列使用样式

data.style.apply(hilight_max, subset=['c', 'e'])
```

```python
# 对指定的区域使用样式

data.style.applymap(color_neg_red, 
                   subset = pd.IndexSlice[:, ['d', 'e', 'g']])
```

```python
# 可以使用 dataframe.style.format 进行格式化处理

data.style.applymap(color_neg_red, subset = 
                    pd.IndexSlice[:, ['d', 'e', 'g']]).format({'b':'{:.2%}', 'e':'{:.2%}', 'f':'{:.2f}'})
```

```python
# 对空值和最值。style中有内置函数进行处理
data.style.highlight_null(null_color='blue')
```

```python
data.style.highlight_max(axis=0)
```

```python
# 使用 seaborn 进行可视化

import seaborn as sns

cm = sns.light_palette("green", as_cmap=True)

s = data.style.background_gradient(cmap=cm)
s
```

```python
# 使用柱状图

data.style.bar(subset=['g'], color='#b65f5f', align='mid')
```

```python
data.style.bar(subset=['g'], color=['#b65f5f','#5fba7d'], align='mid')
```

```python
# 通过样式设置也可以隐藏索引或者一些列
data.style.hide_index()
```

```python
data.style.hide_columns(['f'])
```

```python
# 最后，将处理好的表格导出
data.style.applymap(color_neg_red, subset = 
                    pd.IndexSlice[:, ['d', 'e', 'g']]).\
                    format({'b':'{:.2%}', 'e':'{:.2%}', 'f':'{:.2f}'}).highlight_max(axis=0).to_excel('styled.xlsx', engine='openpyxl')
```

```python
# 这样看如果需要实现百分数的格式化导出，目前最可靠的办法是通过 apply 先进行格式化

data['b'] = data['b'].apply(lambda x:format(x, '.2%'))
data
```
